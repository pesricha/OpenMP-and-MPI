```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

data = {
    'Processes': [1, 2, 4, 8, 16, 32, 64],
    'Avg Time (ms)': [2.33499, 1.24272, 0.633456, 0.320808, 0.16852, 0.0927677, 4.96164]
}

df = pd.DataFrame(data)

df.columns
```

```python
df['Speedup'] = df.iloc[0, 1] / df['Avg Time (ms)']
df
```

```python
sns.set_theme(style="whitegrid")
sns.set_palette("viridis")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

sns.barplot(x='Processes', y='Avg Time (ms)', data=df, ax=ax1, palette='viridis')
ax1.set_title('Average Time vs Processes (Only Finding Element)')
ax1.set_xlabel('Processes')
ax1.set_ylabel('Avg Time (ms)')


sns.barplot(x='Processes', y='Speedup', data=df, ax=ax2, palette='viridis')
ax2.set_title('Speedup vs Processes')
ax2.set_xlabel('Processes')
ax2.set_ylabel('Speedup')

fig.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

data = {
    'Processes': [1, 2, 4, 8, 16, 32, 64],
    'Element Finding Time': [2.33499, 1.24272, 0.633456, 0.320808, 0.16852, 0.0927677, 4.96164],
    'Data Distribution Time': [0.110932, 0.211547, 0.361704, 0.408449, 0.640202, 4.17877, 4.83257],
    'Total Time' : [2.03153, 1.42757, 0.99744, 0.738602, 0.714434, 1.48184, 7.44137]
}

df = pd.DataFrame(data)

df.columns
```

```python
import pandas as pd

# Data
data = {
    "Processes": [1, 2, 4, 8, 16, 32, 64],
    "Avg Distribution Time (ms)": [0.101747, 0.123156, 0.290211, 0.260394, 0.325651, 0.549068, 0.577132],
    "Avg Finding Time (ms)": [2.22195, 1.24738, 0.628362, 0.31729, 0.163942, 0.170848, 0.420712]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display DataFrame
print(df)
```

```python
df['Avg Total Time (ms)'] = df['Avg Distribution Time (ms)'] + df['Avg Finding Time (ms)']
df
```

```python
df['Speedup Distribution'] = df.iloc[0,1]/df['Avg Distribution Time (ms)']
df['Speedup (Finding)'] = df.iloc[0,2]/df['Avg Finding Time (ms)']
df['Speedup (Total)'] = df.iloc[0,3]/df['Avg Total Time (ms)']
df
```

```python
# Set the theme and palette
sns.set_theme(style="whitegrid")
sns.set_palette("viridis")

# Create a new figure
fig, ax = plt.subplots(figsize=(14, 6))

# Plot the data
df_melted = df.melt(id_vars='Processes', value_vars=['Avg Distribution Time (ms)', 'Avg Finding Time (ms)', 'Avg Total Time (ms)'], var_name='Type', value_name='Time (ms)')
sns.barplot(x='Processes', y='Time (ms)', hue='Type', data=df_melted, ax=ax, palette='viridis')


# Customize the plot
ax.set_title('Time vs Processes')
ax.set_xlabel('Processes')
ax.set_ylabel('Time (ms)')
ax.grid(axis='y', linestyle=':')

# Show the plot
plt.show()
```

```python
# Create a new figure
fig, ax = plt.subplots(figsize=(14, 6))

# Melt the dataframe for plotting
df_melted_speedup = df.melt(id_vars='Processes', value_vars=['Speedup Distribution', 'Speedup (Finding)', 'Speedup (Total)'], var_name='Type', value_name='Speedup')

# Plot the data
sns.barplot(x='Processes', y='Speedup', hue='Type', data=df_melted_speedup, ax=ax, palette='viridis')

# Add a horizontal line for max speedup
ax.axhline(y=4.746181, color='black', linestyle='--', label='Max Speedup (4.746181) at 16 processes')
ax.legend()

# Customize the plot
ax.set_title('Speedup vs Processes')
ax.set_xlabel('Processes')
ax.set_ylabel('Speedup')
ax.grid(axis='y', linestyle=':')

# Show the plot
plt.show()
```
